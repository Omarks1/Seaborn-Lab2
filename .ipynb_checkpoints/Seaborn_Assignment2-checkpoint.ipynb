```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Q1: Import seaborn library

```python
# write your code here ^_^
import seaborn as sns
```

```python
tips_dataset = sns.load_dataset('tips')
```

```python
df=tips_dataset
df
```

# Q2: Create a histogram, where the x-axis represents the tip column
- Change the style to 'whitegrid'.
- Split the data out by the day column's categories. Hint: use hue argument.
- Handle the overlapping columns by switching the element argument over to step.

```python
# write your code here ^_^

plt.figure(figsize=(15,8))
sns.histplot(
    df,
    x="tip",
    hue="day",
    multiple="stack",
    element= 'step',
    alpha=0.8);
```

## Q2.1 Write your top three findings 
- Obvious that Thuresday has most number of people paying tips.
- Sunday is the lowest on number of people and tips paied.
- It is obivous that people are paying more than ever on saturday.

# Q3: [Optional] Create a displot, where the x-axis represents the day column
- Split the data out by the sex column's categories. Hint: use hue argument.
- Handle the overlapping columns by switching the multiple argument over to 'dodge'.
- Shrink the bars down using shrink argument to have a nice gap between each of them.
- Change the bars color to 'Set1'.
- Reset the hue order to ['Female', 'Male'].

```python
# write your code here ^_^
 
sns.displot(data=df, x="day", hue='sex', multiple="dodge", shrink=0.3, color='Set1');

# Here is with the rest

sns.displot(data=df, x="day", hue='sex' ,hue_order=['Female','Male'], multiple="dodge", shrink=0.3)
plt.show()
```

## Q3.1 Write your top three findings 
- Male are paying more on sun more than ever.
- Female are doing good on thursday better than other days.
- Female are not paying that much on friday.

# Q4: Create a ecdf plot, where the x-axis represents the tip column
- Change the style to 'darkgrid'.
- Change the font scale to 1.25.
- Split the data out by the time column's categories. Hint: use hue argument.
- Change the lines color to 'summer'.
- Change the line width to 3.

```python
# write your code here ^_^
# a very clear code isn't?

sns.set(font_scale=1.25)
sns.set_style('darkgrid')
sns.ecdfplot(data=df, x="tip", hue='time',color= 'summer',lw=3);
```

## Q4.1 Write your top three findings 
- Number of people who are paying in lunch is much larger than in dinner .
- Tips are paied more on Lunch than Dinner.
- It is obvious that the lunch line had shifted to become more on the number of people. 

# Q5: [Optional] Create a box plot, where the x-axis represents the day column and the y-axis represents the tip column
- Change the style to 'whitegrid'.
- Split the data out by the sex column's categories. Hint: use hue argument.
- Change the bars color to 'Paired'.
- Reset the hue order to ['Female', 'Male'].
- Change the line width to 2.5.

```python
# write your code here ^_^

sns.set_style('whitegrid')
sns.boxplot(x="day", y="tip", hue='sex',data=df, palette='Paired', linewidth=2.5);
```

## Q5.1 Write your top three findings 
- The boxplot shows how tip amounts vary across different days of the week.
- The boxplot highlights the potential of paying the tip amounts between males and females.
- It illustrates the identification of significant differences in tip amounts based on the specific day of the week.

# Q6: Create a violin plot, where the x-axis represents the time column and the y-axis represents the tip column
- Split the data out by the sex column's categories. Hint: use hue argument.
- Split the violin so that each violin contains both sex column's categories. Hint: use split argument.
- Change the shape of the inner part of the violin to 'quartiles'.
- Change the violin color to 'Pastel1'.
- Reset the hue order to ['Female', 'Male'].
- Change the line width to 2.
- Change the legend's location to The upper left.

```python
# write your code here ^_^

plt.figure(figsize=(10,8))

sns.violinplot(
    data=df, 
    x="time", 
    y="tip", 
    hue="sex",
    split= True,
    inner='quartiles',
    paleete='Pastel1', # color
    linewidth=2
);
# legend
plt.legend(loc='upper left');
```

## Q6.1 Write your top three findings 
- Firstly, We can see that male are paying more than girls .
- Secondly, it clarifies that male are paying more on dinner meal than lunch.
- Thirdly, it shows that female have the same attitude of male on paying more on dinner than lunch.


# Q7: [Optional] Create a swarm plot, where the x-axis represents the time column and the y-axis represents the total_bill column

# Q8: Create a correlation heatmap 
- Adjust the figure's size to be (8,5).
- Change the color map to 'RdBu'.
- Add annotations to the heatmap.
- Set the minimum value to be -1.
- Set the maximum value to be 1.

```python
# write your code here ^_^

# sizing the figure
plt.figure(figsize=(8,5)) 

# correlation_matrix = df_selected.corr()
correlation_matrix = df.corr()

# Generating the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='RdBu', vmin=-1, vmax=1, )
plt.title('Correlation Heatmap')
plt.show()
```

## Q8.1 Write your top three findings 
- The heatmap displays the correlation between different variables, providing insights into the strength and direction.
- The heatmap visually represents the correlation matrix, enabling a comprehensive understanding of the relationships between variables.
- The annotated heatmap, with color mapping and numerical values, enables a detailed interpretation of the correlation between variables.
